```
%%cu
#include <iostream>
#






const int row = 20;
const int col = 20;
const int col_row = 40;
const int sizeC = row * col * sizeof (int);
const int size = row * col_row * sizeof (int);

int show(int *matris){
     
    for (int i=0; i<row; i++){
        for (int j=0; j<col; j++){
          std::cout << matris[i + col * j] << "  ";
      }
        std::cout << " ";
        printf("\n");
        //std::cout << "Hello world\n";
    }
    
    return 0;
}



__global__ void matrixMulGPU( int * a, int * b, int * c )
{
    
    int r = threadIdx.y + blockIdx.y * blockDim.y;
    int co = threadIdx.x + blockIdx.x * blockDim.x;
    
    int val = 0;
    if (r < row && co < col) {
      for (int i = 0; i < row; ++i) {
         val += a[r * row + i] * b[i * col_row + co];
       }
    
      c[r * row + co] = val;
    }
    
}

__global__ void mul_matrix( int * a, int * b, int * c )
{
    
    int row1 = threadIdx.y;
    int col1 = threadIdx.x;
    
    int val = 0;
    
    for (int i = 0; i < row; ++i) {
         val += a[row1 + row * i] * b[i + col1 * col_row];
    }
    
    c[row1 * row + col1] = val;
    
    
}


int main() {
    //std::cout << "Hello world\n";
    
    
    int a[row * col_row], b[col_row * row], c[row * col];
    
    for (int i=0; i<row; i++){
        for (int j=0; j<col_row; j++){
           a[i + row * j] = 1;
      }
    }
    
    for (int i=0; i<col_row; i++){
        for (int j=0; j<row; j++){
           b[i + col_row * j] = 1;
      }
    }
    
     for (int i=0; i<col; i++){
        for (int j=0; j<row; j++){
           c[i + row * j] = -1;
      }
    }
    
    
    int *d_a, *d_b, *d_c;
    cudaEvent_t start, stop;
    
    int size1 = col * col_row * sizeof(int);
    int size2 = col_row * row * sizeof(int);
    int size3 = col * row * sizeof(int);
    
    cudaMalloc ((void **) &d_a, size1);
    cudaMalloc ((void **) &d_b, size2);
    cudaMalloc ((void **) &d_c, size3);
    
    cudaMemcpy(d_a, a, size1, cudaMemcpyHostToDevice);
    cudaMemcpy(d_b, b, size2, cudaMemcpyHostToDevice);
    //cudaMemcpy(d_c, c, size3, cudaMemcpyHostToDevice);
    
    //dim3 threads_per_block(32, 32, 1);
    //dim3 number_of_blocks(col / threads_per_block.x + 1, col_row / threads_per_block.y + 1, 1);
    
    cudaEventCreate(&start);
    cudaEventCreate(&stop);
    cudaEventRecord(start, 0);
    
    dim3 dimGrid(1,1);
    dim3 dimBlock(col, row);
    mul_matrix <<< dimGrid, dimBlock >>> ( d_a, d_b, d_c );
    
    
    cudaEventRecord(stop, 0);
    cudaEventSynchronize(stop);
    
    
    cudaMemcpy(c, d_c, size3, cudaMemcpyDeviceToHost);
    float elapsedTime;
    cudaEventElapsedTime(&elapsedTime, start, stop);
    
    
    show(c);
    
    printf("\n\nTime:  %3.5f ms\n", elapsedTime);
    
    
   
    
    
  //  int a[20][40], b[40][20], c[20][20];
  //  
  //  
  //  cudaMallocManaged (&a, size);
  //  cudaMallocManaged (&b, size);
  //  cudaMallocManaged (&c, sizeC);
  //  //cudaMallocManaged (&c_gpu, size);
  //  
  //  for (int i=0; i<40; i++){
  //      for (int j=0; j<40; j++){
  //         a[i][j] = 1;
  //         b[i][j] = 2;
  //    }
  //  }
  //  
  //  dim3 threads_per_block(32, 32, 1);
  //  dim3 number_of_blocks(N / threads_per_block.x + 1, N / threads_per_block.y + 1, 1);
  //  matrixMulGPU <<< number_of_blocks, threads_per_block >>> ( a, b, c_gpu );
  //  
  //  show(c);
    
    return 0;
}

```
